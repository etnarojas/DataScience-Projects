```python
import plotly.plotly as py
import plotly.graph_objs as go
trace0 = go.Scatter(
    y = BitcoinData['Open'],
    x = BitcoinData['Date'],
    mode='markers',
    name = 'OPEN',
    marker = dict(
        size = 4,
        color = 'rgba(10, 255, 31, 1)',
        line = dict(
            width = 1,
        )
    )
    
   
   
)
trace1 = go.Scatter(
    y = BitcoinData['High'],
    x = BitcoinData['Date'],
    name = 'HIGH',
    mode = 'markers',
    marker = dict(
        size = 3,
        color = 'rgba(0, 32, 240, 1)',
        line = dict(
            width = 1,
        )
    )
    
   
)
trace2 = go.Scatter(
    y = BitcoinData['Low'],
    x = BitcoinData['Date'],
    mode = 'markers',
    name = 'LOW',
    marker = dict(
        size = 2,
        color = 'rgba(255, 10, 10, 1)',
        line = dict(
            width = 1,
        )
    )
    
   
)
trace3 = go.Scatter(
    y = BitcoinData['Close'],
    x = BitcoinData['Date'],
    mode = 'markers',
    name = 'CLOSE',
    marker = dict(
        size = 1,
        color = 'rgba(255, 122, 196, .9)',
        line = dict(
            width = 1,
        )
    )
    
   
)



data = [trace0,trace1,trace2,trace3]
py.iplot(data, filename='scatter-mode')
```

```python
YearD = pd.ExcelFile('YearData.xlsx')
Ydata = pd.read_excel(YearD, 'Sheet1')
```

```python
Ydata.head(5)
```

```python
trace1 = go.Bar(
    x=Ydata['Date'],
    y=Ydata['min'],
    name='Minimum Value'
)
trace2 = go.Bar(
    x=Ydata['Date'],
    y=Ydata['max'],
    name='Maximim Value'
)

data = [trace1, trace2]
layout = go.Layout(
    barmode='group'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```
