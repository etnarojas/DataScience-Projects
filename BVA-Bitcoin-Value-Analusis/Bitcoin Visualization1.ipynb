```python

trace1 = go.Scatter(
    y = Bitcoin['Price'],
    x = Bitcoin['Date'],
    mode='markers',
    marker=dict(
        size='1.5',
        color = np.random.randn(59865), #set color equal to a variable
        colorscale='Viridis'
        
    )
)
data = [trace1]
layout = go.Layout(
    title='Date vs. Bitcoin $ Price',
    xaxis=dict(
        title='Years',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Price in $',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )

)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='scatter-plot-with-colorscale')
```

```python
HIGH= go.Scatter(
    name='lower_bound',
    x=Bitcoin['Date'],
    y=Bitcoin['High'],
    mode='lines',
    marker=dict(color="444"),
    line=dict(width=0),
    fillcolor='rgba(68, 68, 68, 0.3)',
    fill='tonexty')
PRICE = go.Scatter(
    name='trace',
    x=Bitcoin['Date'],
    y=Bitcoin['Price'],
    mode='lines',
    line=dict(color='rgb(31, 119, 180)'),
    fillcolor='rgba(68, 68, 68, 0.3)',
    fill='tonexty')

LOW = go.Scatter(
    name='Lower Bound',
    x=Bitcoin['Date'],
    y=Bitcoin['Low'],
    marker=dict(color="444"),
    line=dict(width=3),
    mode='lines')
data = [HIGH, PRICE, LOW]

layout = go.Layout(
    yaxis=dict(title='Date'),
    showlegend = True)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='pandas-continuous-error-bars')
```
